```python
from imblearn.over_sampling import SMOTE
```

```python
X= df.drop('quality', axis=1)
y= df['quality']

sampling_strategy = {
    3: 240,
    8: 220,
    4: 270,
    7: 310
    
}

oversample = SMOTE(sampling_strategy=sampling_strategy, random_state=42)
X_resampled, y_resampled = oversample.fit_resample(X, y)
```

```python
df1 = pd.DataFrame(X_resampled, columns=X.columns)
df2 = pd.DataFrame(y_resampled, columns=['quality'])
```

```python
oversampled_df = pd.concat([df1, df2], axis=1)
```

```python
oversampled_df.head()
```

```python
oversampled_df['quality'].value_counts()
```

```python
oversampled_df['quality'].value_counts().plot(kind='bar')
```
