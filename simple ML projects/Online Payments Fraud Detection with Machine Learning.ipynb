```python
df['type'] = df['type'].map({'PAYMENT':2, 'TRANSFER':4, 'CASH_OUT':1, 'DEBIT':5, 'CASH_IN':3})
df['isFraud'] = df['isFraud'].map({0:'No Fraud', 1:'Fraud'})
df.head()
```

train a classification model to classify fraud and non-fraud transactions

```python
#splitting the data
X = np.array(df[['type', 'amount', 'oldbalanceOrg', 'newbalanceOrig']])
y = np.array(df['isFraud'])
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
print(model.score(X_test, y_test))
```

classify whether a transaction is a fraud or not by feeding about a transaction into the model:


```python
#predict
#type', 'amount', 'oldbalanceOrg', 'newbalanceOrig'

prediction_features = np.array([[4, 9000.60, 9000.60, 0.0]])
print(model.predict(prediction_features))
```
